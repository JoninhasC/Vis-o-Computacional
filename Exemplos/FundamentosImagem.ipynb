```python
import cv2
import matplotlib.pyplot as plt
```

Exemplo Pr√°tico: Abrindo e Exibindo uma Imagem

```python
# Carregar a imagem
imagem = cv2.imread("gato.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: N√£o foi poss√≠vel carregar a imagem.")
else:
    # Converter de BGR para RGB (para exibi√ß√£o correta no matplotlib)
    imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)

    # Exibir a imagem no Jupyter Notebook
    plt.imshow(imagem_rgb)
    plt.axis('off')  # Remove os eixos para uma exibi√ß√£o mais limpa
    plt.show()
```

```python
# Carregar a imagem
imagem = cv2.imread("gato.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: N√£o foi poss√≠vel carregar a imagem.")
else:
    # Obter altura, largura e canais
    altura, largura, canais = imagem.shape

    # Obter a profundidade de cor
    profundidade_cor = imagem.dtype

    # Exibir informa√ß√µes
    print(f"Dimens√£o: {largura * altura} pixels")
    print(f"Resolu√ß√£o: {largura}x{altura}")
    print(f"N√∫mero de canais: {canais}")
    print(f"Profundidade de cor: {profundidade_cor}")
```

Exemplo Pr√°tico:Convertendo para Tons de Cinza

```python
# Carregar a imagem
imagem = cv2.imread("gato.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: N√£o foi poss√≠vel carregar a imagem.")
else:
    # Converter para escala de cinza
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Obter a profundidade de cor
    profundidade_cor = imagem_cinza.dtype
    print("Profundidade de cor:", profundidade_cor)

    # Exibir imagem dentro do Jupyter
    plt.imshow(imagem_cinza, cmap='gray')
    plt.axis('off')  # Remove os eixos para visualiza√ß√£o mais limpa
    plt.show()
```

Exemplo Pr√°tico: Acessando um Pixel Espec√≠fico

```python
# Carregar a imagem
imagem = cv2.imread("gato.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: N√£o foi poss√≠vel carregar a imagem.")
else:
    # Converter para escala de cinza
    imagem_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

    # Definir coordenadas do recorte (y1:y2, x1:x2)
    x1, y1 = 50, 50   # Canto superior esquerdo
    x2, y2 = 200, 200  # Canto inferior direito

    # Verificar se as coordenadas est√£o dentro da imagem
    altura, largura = imagem_cinza.shape
    if 0 <= x1 < largura and 0 <= x2 <= largura and 0 <= y1 < altura and 0 <= y2 <= altura:
        # Recortar a imagem
        imagem_recortada = imagem_cinza[y1:y2, x1:x2]

        # Exibir a parte recortada
        plt.imshow(imagem_recortada, cmap='gray')
        plt.axis('off')  # Remove os eixos para uma visualiza√ß√£o mais limpa
        plt.show()
    else:
        print("Erro: Coordenadas do recorte est√£o fora dos limites da imagem.")
```
