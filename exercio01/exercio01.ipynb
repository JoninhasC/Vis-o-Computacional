```python
import cv2
import matplotlib.pyplot as plt
```

Baixe a imagem vermelho.jpg no moodle, e modifique as cores do pixels da imagem para a
cor preta e salve como preta.jpg

```python
# Carregar a imagem
imagem = cv2.imread("vermelho.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: Não foi possível carregar a imagem.")
else:
    # Extrair o canal vermelho da imagem
    canal_vermelho = imagem[:, :, 2]  # Canal vermelho (BGR -> R)

    # Aplicar limiarização invertida (thresholding invertido)
    limiar = 100  # Definir o valor do limiar (threshold)
    # O valor do limiar pode ser ajustado conforme necessário
    # A função cv2.threshold retorna dois valores: o primeiro é o 
    # limiar aplicado e o segundo é a imagem binária resultante
    _, imagem_preto = cv2.threshold(canal_vermelho, limiar, 255, cv2.THRESH_BINARY_INV) # Aplica o limiar invertido

    # Salvar a imagem convertida
    cv2.imwrite("preta.jpg", imagem_preto) # Salva a imagem convertida como 'preta.jpg'
    # Exibir mensagem de sucesso
    print("Imagem convertida para preto salva como 'preta.jpg'.")

    # Exibir a imagem original e a imagem convertida
    plt.imshow(imagem_preto, cmap='gray')
    plt.axis('off')  
    plt.show()
```

Abra a imagem preta.jpg e desenhe um quadrado branco (50x50 pixels) no centro da
imagem e salve como quadradobranco50.jpg.

```python
# Carregar a imagem
imagem = cv2.imread("preta.jpg")
# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: Não foi possível carregar a imagem.")
else:
    # Obter as dimensões da imagem
    altura, largura = imagem.shape[:2]

    # Calcular as coordenadas do quadrado branco
    centro_x, centro_y = largura // 2, altura // 2 
    inicio_x, inicio_y = centro_x - 25, centro_y - 25
    fim_x, fim_y = centro_x + 25, centro_y + 25

    # Desenhar o quadrado branco na imagem
    cv2.rectangle(imagem, (inicio_x, inicio_y), (fim_x, fim_y), (255, 255, 255), -1)

    # Salvar a imagem modificada
    cv2.imwrite("quadradobranco50.jpg", imagem)
    print("Imagem com quadrado branco salva como 'quadradobranco50.jpg'.")

    # Exibir a imagem modificada
    plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

Abra a imagem preta.jpg e desenhe um quadrado branco (25x25 pixels) no centro da
imagem e salve como quadradobranco25.jpg.


```python
# Carregar a imagem
imagem = cv2.imread("preta.jpg")
# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: Não foi possível carregar a imagem.")
else:
    # Obter as dimensões da imagem
    altura, largura = imagem.shape[:2]

    # Calcular as coordenadas do quadrado branco
    centro_x, centro_y = largura // 2, altura // 2 
    inicio_x, inicio_y = int(centro_x - 12.5), int(centro_y - 12.5)
    fim_x, fim_y = int(centro_x + 12.5), int(centro_y + 12.5)

    # Desenhar o quadrado branco na imagem
    cv2.rectangle(imagem, (inicio_x, inicio_y), (fim_x, fim_y), (255, 255, 255), -1)

    # Salvar a imagem modificada
    cv2.imwrite("quadradobranco25.jpg", imagem)
    print("Imagem com quadrado branco salva como 'quadradobranco50.jpg'.")

    # Exibir a imagem modificada
    plt.imshow(cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

Subtraia a imagem quadradobranco25.jpg de quadradobranco50.jpg e salve o resultado
como subtracao.jpg.


```python
# Carregar as imagens
imagem1 = cv2.imread("quadradobranco50.jpg")
imagem2 = cv2.imread("quadradobranco25.jpg")

# Verificar se as imagens foram carregadas corretamente
if imagem1 is None or imagem2 is None:
    print("Erro: Não foi possível carregar uma ou ambas as imagens.")
else:
    # Subtrair as imagens
    subtracao = cv2.subtract(imagem1, imagem2)

    # Salvar o resultado da subtração
    cv2.imwrite("subtracao.jpg", subtracao)
    print("Resultado da subtração salvo como 'subtracao.jpg'.")

    # Exibir a imagem resultante
    plt.imshow(cv2.cvtColor(subtracao, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

Pegue a imagem subtracao.jpg, inverta suas cores, e aplique um XOR com a imagem
quadradobranco50.jpg.

```python
# Carregar as imagens
imagem_subtracao = cv2.imread("subtracao.jpg")
imagem_quadrado = cv2.imread("quadradobranco50.jpg")

# Verificar se as imagens foram carregadas corretamente
if imagem_subtracao is None or imagem_quadrado is None:
    print("Erro: Não foi possível carregar uma ou ambas as imagens.")
else:
    # Inverter as cores da imagem subtracao.jpg
    imagem_invertida = cv2.bitwise_not(imagem_subtracao)

    # Aplicar a operação XOR entre a imagem invertida e quadradobranco50.jpg
    resultado_xor = cv2.bitwise_xor(imagem_invertida, imagem_quadrado)

    # Salvar o resultado da operação XOR
    cv2.imwrite("resultado_xor.jpg", resultado_xor)
    print("Resultado da operação XOR salvo como 'resultado_xor.jpg'.")

    # Exibir a imagem resultante
    plt.imshow(cv2.cvtColor(resultado_xor, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

Baixe e abra a imagem televisao.jpg e recorte a televisão presente na imagem removendo
sua base. Depois inverta a imagem horizontalmente. Dica: a televisão tem cerca de 190
pixels de altura e 270 pixels de largura.


```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread("televisao.jpg")

# Verificar se a imagem foi carregada corretamente
if imagem is None:
    print("Erro: Não foi possível carregar a imagem.")
else:
    # Definir as dimensões do recorte (removendo a base)
    altura_televisao = 220# Altura da televisão
    largura_televisao = 330 # Largura da televisão
    recorte = imagem[30:altura_televisao,60:largura_televisao]  # Recortar a televisão

    # Inverter a imagem horizontalmente
    imagem_invertida = cv2.flip(recorte, 1)

    # Salvar a imagem resultante
    cv2.imwrite("televisao_invertida.jpg", imagem_invertida)
    print("Imagem da televisão recortada e invertida salva como 'televisao_invertida.jpg'.")

    # Exibir a imagem resultante
    plt.imshow(cv2.cvtColor(imagem_invertida, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
